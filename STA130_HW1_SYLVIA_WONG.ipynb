It is crucial to use del df['col'] for columns with high proportions of missing data first because, as seen in the example above, using df.dropna() first would remove every single customer from the data set because every entry has at least one missing value in the "useless columns". This significantly decreases the sample size. However, this is not ideal because the name and email of each customer are not missing and should be retained.
````python
# Part 4: After (note: I used a for loop because I made my code general to work with any dataset)

import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
df = pd.read_csv(url)

# Find Missing Values Per Column
missing_value_per_column = df.isnull().sum()

# Remove Useless Columns
for col in df.columns:
    if (missing_value_per_column[col] / len(df)) > 0.5: # If more than a half of the column is missing, it is removed (the threshold is 50%)
        del (df[col])
    
# Remove Rows with Missing Values
df_cleaned = df.dropna()
print (df_cleaned)
````

In my approach, I first counted the number of missing values per column. Then, using a for loop, I looped through each column in the dataset and compared the number of missing values in that column to the total number of values in that column. If more than 50% of the values in the column are missing, that column is removed. Then I removed the remaining rows with missing values. 

I used this approach because it minimizes the amount of data removed and retains as much useful information as possible. By checking the proportion of missing values to total data in each column, I can weigh the importance of keeping the column or keeping the entries that contain those missing values. In this case, the "deck" column had more than 50% missing values. Removing the deck column would outweigh removing all the entries with missing values in the deck column because if I were to remove the entries, I would end up removing over 1/2 of the data. This would significantly cut down the size of the data set, giving less to analyze. Although the threshold of "50%" could vary, I chose 50% because I believe that if a column is more than 50% empty, it does not convey too much important information. Furthermore, if you instead used df.dropna(), you would cut down the data set by more than 1/2, resulting in a significantly smaller sample size.

From before, the columns "age", "embarked", "embark_town" and "deck" have missing values. Age, embarked, and embark_town were not removed because the missing values in these column make up less than 50% of the column. This indicates that these column are important enough to keep. Instead, the entries with missing values in the "age", "embark_town", and "embark" column are removed. This would minimize the data removed.

### **Summary of Interactions**

#### 1. **Understanding `del df['col']` vs. `df.dropna()`**

- **`del df['col']`**:
  - Removes a specific column entirely from the DataFrame.
  - Preferred when you want to remove irrelevant, redundant, or highly missing columns.
  
- **`df.dropna()`**:
  - Removes rows or columns containing missing values based on specified conditions.
  - Use when cleaning the DataFrame by removing rows or columns with missing values without targeting a specific column.

- **Use Case Comparison**:
  - **`df.dropna()`** is better when you want to remove incomplete rows or columns to maintain overall data integrity.
  - **`del df['col']`** is preferred when a specific column has too many missing values or is irrelevant.

#### 2. **Using `del df['col']` and `df.dropna()` Together**

- **Scenario**: First, remove unnecessary columns with many missing values using `del df['col']`, then use `df.dropna()` to further clean rows or columns with remaining missing values.
  
- **Example**:
  ```python
  import pandas as pd
  
  data = {
      'CustomerID': [1, 2, 3, 4, 5],
      'Email': ['alice@example.com', 'bob@example.com', None, 'david@example.com', 'eva@example.com'],
      'PhoneNumber': [None, None, '123-456-7890', None, None],
      'Notes': [None, 'Called back', None, 'Preferred customer', None]
  }
  
  df = pd.DataFrame(data)
  
  # Remove 'PhoneNumber' and 'Notes' columns due to high missing values
  del df['PhoneNumber']
  del df['Notes']
  
  # Remove rows with missing values
  df_cleaned = df.dropna()
  ```

#### 3. **Counting Values in a Column**

- **Methods**:
  - `len(df['col'])`: Counts all values, including missing values.
  - `df['col'].count()`: Counts only non-missing values.
  - `df['col'].value_counts()`: Counts occurrences of each unique value, excluding missing values.

#### 4. **Handling the KeyError: 'col'**

#### 5. **Determining When a Column Has Too Many Missing Values**

- **Factors to Consider**:
  - **Threshold**: Commonly used thresholds are 50%, 60-80%, and above 80% missing values.
  - **Importance**: Essential columns may be retained even with missing data.
  - **Data Imputability**: Consider if missing values can be imputed reliably.
  - **Correlation**: Columns highly correlated with others can sometimes be imputed.

- **Example**:
  ```python
  # Calculate the percentage of missing values
  missing_percentage = df.isnull().mean() * 100
  
  # Define a threshold for dropping columns
  threshold = 50
  columns_to_drop = missing_percentage[missing_percentage > threshold].index
  
  # Drop columns exceeding the missing values threshold
  df_cleaned = df.drop(columns=columns_to_drop)
  ```

- **Guidelines**:
  - Use threshold-based criteria to remove columns with excessive missing values.
  - Balance between data retention and the quality of remaining data, focusing on the impact on analysis.
