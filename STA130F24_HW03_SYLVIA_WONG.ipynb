````python
import pandas as pd
import plotly.graph_objects as go

# Load dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
penguins = pd.read_csv(url)

# Group by species and compute stats
stats = penguins.groupby('species')['flipper_length_mm'].agg(
    mean='mean',
    median='median',
    min_val='min',
    max_val='max',
    q1=lambda x: x.quantile(0.25),
    q3=lambda x: x.quantile(0.75),
    std='std'
).reset_index()

# Add range defined by two standard deviations away from the mean
stats['lower_2std'] = stats['mean'] - 2 * stats['std']
stats['upper_2std'] = stats['mean'] + 2 * stats['std']

# Define color schemes for each species
color_schemes = {
    'Adelie': {
        'mean': 'black',      # Mean in black
        'median': 'red',      # Median in red
        'range': 'rgba(86, 180, 233, 0.2)',   # Light Blue
        'iqr': 'rgba(233, 153, 51, 0.3)',      # Light Orange
        'std_dev': 'rgba(0, 158, 73, 0.2)'     # Light Green
    },
    'Chinstrap': {
        'mean': 'black',
        'median': 'red',
        'range': 'rgba(86, 180, 233, 0.2)',
        'iqr': 'rgba(233, 153, 51, 0.3)',
        'std_dev': 'rgba(0, 158, 73, 0.2)'
    },
    'Gentoo': {
        'mean': 'black',
        'median': 'red',
        'range': 'rgba(86, 180, 233, 0.2)',
        'iqr': 'rgba(233, 153, 51, 0.3)',
        'std_dev': 'rgba(0, 158, 73, 0.2)'
    }
}

# Create individual figures for each species
for species in stats['species']:
    fig = go.Figure()
    
    species_data = penguins[penguins['species'] == species]['flipper_length_mm'].dropna()  # Drop NaN values
    
    # Add histogram for the current species
    fig.add_trace(go.Histogram(
        x=species_data,
        name=f'{species} flipper_length_mm',
        opacity=0.6,
        showlegend=False
    ))
    
    # Get statistics for the current species
    species_stats = stats[stats['species'] == species]
    
    mean_val = species_stats['mean'].values[0]
    median_val = species_stats['median'].values[0]
    min_val = species_stats['min_val'].values[0]
    max_val = species_stats['max_val'].values[0]
    q1_val = species_stats['q1'].values[0]
    q3_val = species_stats['q3'].values[0]
    lower_2std = species_stats['lower_2std'].values[0]
    upper_2std = species_stats['upper_2std'].values[0]

    # Get colors for the current species
    colors = color_schemes[species]

    # Add vertical lines for mean and median
    fig.add_vline(x=mean_val, line_dash="dash", line_color=colors['mean'], 
                  name='Mean', legendgroup=species, showlegend=True)
    fig.add_vline(x=median_val, line_dash="dashdot", line_color=colors['median'], 
                  name='Median', legendgroup=species, showlegend=True)

    # Add vertical rectangles for range, IQR, and two std range
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor=colors['range'], line_width=0,
                  name='Range', legendgroup=species, showlegend=True)
    fig.add_vrect(x0=q1_val, x1=q3_val, fillcolor=colors['iqr'], line_width=0,
                  name='IQR', legendgroup=species, showlegend=True)
    fig.add_vrect(x0=lower_2std, x1=upper_2std, fillcolor=colors['std_dev'], line_width=0,
                  name='2 Std Dev Range', legendgroup=species, showlegend=True)

    # Update layout for the current figure
    fig.update_layout(
        title=f"{species} Flipper Length Distribution with Marked Statistics",
        xaxis_title="Flipper Length (mm)",
        yaxis_title="Count",
        barmode='overlay',
        height=600,
        showlegend=True
    )

    # Show the figure for the current species
    fig.show(renderer="png")
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import gaussian_kde

# Load dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
penguins = pd.read_csv(url)

# Group by species and compute stats
stats = penguins.groupby('species')['flipper_length_mm'].agg(
    mean='mean',
    median='median',
    min_val='min',
    max_val='max',
    q1=lambda x: x.quantile(0.25),
    q3=lambda x: x.quantile(0.75),
    std='std'
).reset_index()

# Add range defined by two standard deviations away from the mean
stats['lower_2std'] = stats['mean'] - 2 * stats['std']
stats['upper_2std'] = stats['mean'] + 2 * stats['std']

# Set improved color scheme with black IQR and lighter green for standard deviation
color_schemes = {
    'Adelie': {
        'mean': 'black',
        'median': 'red',
        'range': (1, 0.5, 0.5, 0.3),  # Darker pink with 30% opacity
        'iqr': (0, 0, 0, 0.4),          # Black with 40% opacity
        'std_dev': (0.0, 0.8, 0.0, 0.4)  # Lighter green with 40% opacity
    },
    'Chinstrap': {
        'mean': 'black',
        'median': 'red',
        'range': (1, 0.5, 0.5, 0.3),  # Darker pink with 30% opacity
        'iqr': (0, 0, 0, 0.4),          # Black with 40% opacity
        'std_dev': (0.0, 0.8, 0.0, 0.4)  # Lighter green with 40% opacity
    },
    'Gentoo': {
        'mean': 'black',
        'median': 'red',
        'range': (1, 0.5, 0.5, 0.3),  # Darker pink with 30% opacity
        'iqr': (0, 0, 0, 0.4),          # Black with 40% opacity
        'std_dev': (0.0, 0.8, 0.0, 0.4)  # Lighter green with 40% opacity
    }
}

# Create a figure with 3 subplots in a row
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Define species for each subplot
species_list = ['Adelie', 'Chinstrap', 'Gentoo']

# Store the maximum y-axis value to synchronize the y-limits later
max_density = 0

for i, species in enumerate(species_list):
    ax = axes[i]
    
    species_data = penguins[penguins['species'] == species]['flipper_length_mm'].dropna()
    
    # Calculate KDE values manually
    kde = gaussian_kde(species_data)
    x_vals = np.linspace(species_data.min(), species_data.max(), 1000)
    y_vals = kde(x_vals)
    
    # Plot KDE
    ax.fill_between(x_vals, y_vals, color='gray', alpha=0.3)
    
    # Track the maximum y-axis value
    current_max_density = y_vals.max()
    max_density = max(max_density, current_max_density)

    # Get statistics for the current species
    species_stats = stats[stats['species'] == species]
    
    mean_val = species_stats['mean'].values[0]
    median_val = species_stats['median'].values[0]
    min_val = species_stats['min_val'].values[0]
    max_val = species_stats['max_val'].values[0]
    q1_val = species_stats['q1'].values[0]
    q3_val = species_stats['q3'].values[0]
    lower_2std = species_stats['lower_2std'].values[0]
    upper_2std = species_stats['upper_2std'].values[0]
    
    # Plot vertical lines for mean and median
    ax.axvline(mean_val, color=color_schemes[species]['mean'], linestyle='--', label='Mean')
    ax.axvline(median_val, color=color_schemes[species]['median'], linestyle='-.', label='Median')
    
    # Plot rectangles for range, IQR, and 2 standard deviations range
    ax.axvspan(min_val, max_val, color=color_schemes[species]['range'], label='Range')
    ax.axvspan(q1_val, q3_val, color=color_schemes[species]['iqr'], label='IQR')
    ax.axvspan(lower_2std, upper_2std, color=color_schemes[species]['std_dev'], label='2 Std Dev Range')
    
    # Set title for the current subplot
    ax.set_title(f"{species} Flipper Length KDE")
    
    # Add a legend to each panel
    ax.legend()

# Set the same y-axis limit for all subplots
for ax in axes:
    ax.set_ylim(0, max_density * 1.1)  # Add a little buffer to avoid cutting off the peak

# Set common labels
fig.text(0.5, 0.04, 'Flipper Length (mm)', ha='center', va='center', fontsize=14)
fig.text(0.04, 0.5, 'Density', ha='center', va='center', rotation='vertical', fontsize=14)

# Adjust the layout
plt.tight_layout()
plt.show()
````

The mean is the average of the data set while the median is the 50th percentile in the data set. The mean and median are related to right and left skewness in the following way. Because the mean is more affected by outliers and extremely high or low values as it depends on the magnitudes in the data, the mean tends to get "pulled in the direction of the skew" while the median, which is less affected by outliers as it only depends on the central value rather than its magnitude, will remain unaffected. In a symmetrical distribution, the mean and the median are equal but in a right skewed distribution, the mean is pulled to the right (mean > median) while in a left skewed distribution, the mean is pulled to the left (mean < median). This is because, the end of the distribution that has a "tail of extremely high or low values" will cause the mean to shift to that side. 

In the following code, you can see that the histogram has the bulk of its data on the left and the "tail" of extreme values on the right. This represents a right-shifted distribution. As seen by the blue dotted line, a right-shifted ditribution has mean > median because the mean is more affected by the extreme values in the tail on the right, as explained before. 

In the second graph, you can see a left-shifted distribution where the bulk of the data is on the right but the tail of extremely low values is on the left. As you can see, in the left-shifted distribution, the mean is less than the median as it is pulled by the outliers on the left.

````python
from scipy import stats
import plotly.express as px
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Generate the sample from a Gamma distribution
sample1 = stats.gamma(a=2, scale=2).rvs(size=1000)

# Create the histogram
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title="Right-Shifted Graph")

# Calculate the mean and median
mean_value = sample1.mean()
median_value = np.quantile(sample1, 0.5)  # 0.5 quantile is the median

# Convert the figure to a Plotly graph object to add traces
fig1 = go.Figure(fig1)

# Add a line for the mean
fig1.add_vline(x=mean_value, line_dash="dash", line_color="blue", name="Mean", annotation=None)

# Add a line for the median
fig1.add_vline(x=median_value, line_dash="dash", line_color="red", name="Median", annotation=None)

# Add the mean and median lines to the legend by using an invisible scatter trace
fig1.add_trace(go.Scatter(x=[None], y=[None], mode='lines',
                          line=dict(color="blue", dash="dash"),
                          name='Mean'))
fig1.add_trace(go.Scatter(x=[None], y=[None], mode='lines',
                          line=dict(color="red", dash="dash"),
                          name='Median'))

# Show the plot with the legend
fig1.show(renderer="png")

sample2 = -stats.gamma(a=2,scale=2).rvs(size=1000)

# Generate the sample from a negative Gamma distribution
sample2 = -stats.gamma(a=2, scale=2).rvs(size=1000)

# Create the histogram
fig2 = px.histogram(pd.DataFrame({'data': sample2}), x="data", title="Negative Gamma Distribution with Mean and Median")

# Calculate the mean and median
mean_value = sample2.mean()
median_value = np.quantile(sample2, 0.5)  # 0.5 quantile is the median

# Convert the figure to a Plotly graph object to add traces
fig2 = go.Figure(fig2)

# Add a line for the mean
fig2.add_vline(x=mean_value, line_dash="dash", line_color="blue", name="Mean", annotation=None)

# Add a line for the median
fig2.add_vline(x=median_value, line_dash="dash", line_color="red", name="Median", annotation=None)

# Add the mean and median lines to the legend by using an invisible scatter trace
fig2.add_trace(go.Scatter(x=[None], y=[None], mode='lines',
                          line=dict(color="blue", dash="dash"),
                          name='Mean'))
fig2.add_trace(go.Scatter(x=[None], y=[None], mode='lines',
                          line=dict(color="red", dash="dash"),
                          name='Median'))

# Show the plot with the legend
fig2.show(renderer="png")
````
